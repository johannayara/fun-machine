```python
import pandas as pd
from difflib import get_close_matches
```

```python
df_salaire = pd.read_excel('../data/SHS_TM.xlsx',index_col=0, header=1).drop(['Total'], axis=1)
salaire_abrev = df_salaire.columns
```

```python
df_encoded_1923 = pd.read_csv("../data_processing/heatmap/inputs/geocoded_table_1923_clf.csv")
```

```python
def find_closest_job(job):
    matches = get_close_matches(job, salaire_abrev
, n=1, cutoff=0.5)
    if matches:
        col = matches[0]
    else:
        col = job
    return col
df_encoded_1923['job_clf'] = df_encoded_1923['job_clf'].apply(find_closest_job)
```
